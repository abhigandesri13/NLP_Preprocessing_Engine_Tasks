### **Task 1: Conceptual Understanding**

 **1. What is the difference between "Love" and "love" in NLP?**

Ans. In NLP, "Love" and "love" are treated as two different words because text processing systems are case-sensitive.
This means uppercase and lowercase letters are considered different tokens.
However, both words carry the same meaning in real-world language.
To avoid duplication and inconsistency, we usually convert all text into lowercase during preprocessing.

**2. What happens if stopwords are not removed?**

Ans. If stopwords are not removed, the dataset will contain many common words such as "is", "the", and "and".
These words do not contribute much to the actual meaning of the sentence.
As a result, the data becomes larger and noisier, which can reduce model efficiency and increase processing time.
It may also affect the accuracy of machine learning models.

**3. Provide two real-world scenarios where removing stopwords can be harmful.**

Ans. One scenario is sentiment analysis.
For example, in the sentence "I am not happy", removing the word "not" changes the meaning completely and gives the wrong sentiment.                  
                       
Another scenario is in question-answering systems.
Words like "what", "how", and "why" are essential to understand the intent of the question.
Removing them can lead to incorrect interpretation.

**4. Explain the difference between stemming and lemmatization.**

Ans. Stemming is a basic technique that reduces words to their root form by removing suffixes.                                           
However, it may produce incorrect or incomplete words.
For example, "running" may become "runn".                                  
Lemmatization is a more advanced method that converts words into their correct base form using vocabulary and meaning.                                 
For example, "better" becomes "good".
It provides more accurate and meaningful results compared to stemming.

### **Task 2: Build Advanced Preprocessing Function**

```python
import re

def preprocess_text(text):
    # Handle empty or invalid input
    if not text or not isinstance(text, str):
        return [], ""

    # 1. Remove URLs and email-like patterns
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)

    # 2. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 3. Convert to lowercase
    text = text.lower()

    # 4. Handle repeated characters (e.g., soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 5. Remove special characters (keep only alphabets and spaces)
    text = re.sub(r'[^a-z\s]', '', text)

    # 6. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # 7. Remove very short tokens (≤ 2), but keep "no" and "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    # Final cleaned sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

I implemented a function called preprocess_text that cleans raw text step by step.
It removes unwanted elements like numbers, URLs, and special characters.
It also converts text to lowercase, handles repeated characters, and removes unnecessary short words.
Finally, it returns cleaned tokens and a structured sentence.

### **Task 4: Token Analytics**

```python
def token_analysis(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens == 0:
        avg_length = 0
    else:
        avg_length = sum(len(word) for word in tokens) / total_tokens

    return total_tokens, unique_tokens, avg_length


# Apply on test sentences
for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    total, unique, avg_len = token_analysis(tokens)

    print("Original Text:", sentence)
    print("Tokens:", tokens)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg_len, 2))
    print("-" * 50)
```

### **Task 5: Frequency Analysis**

```python
from collections import Counter

all_tokens = []

# Collect tokens from all sentences
for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

# Apply Counter
word_counts = Counter(all_tokens)

# Top 10 most frequent words
top_10 = word_counts.most_common(10)

# Top 5 least frequent words
least_5 = word_counts.most_common()[-5:]

print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(word, ":", count)

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(word, ":", count)
```

I combined all tokens from the processed sentences and used the Counter function to calculate word frequency.
This helps identify commonly used words as well as rare words in the dataset.
It is useful for understanding patterns in text data.


### **Task 6: Build Full Pipeline**

```python
def full_pipeline(text_list):
    # Handle invalid input
    if not text_list or not isinstance(text_list, list):
        return {"tokens": [], "clean_sentences": []}

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print("All Tokens:", result["tokens"])
print("Clean Sentences:", result["clean_sentences"])
```

I created a full_pipeline function that processes multiple sentences at once.
It uses the preprocess_text function for each input sentence.
All tokens are combined into a single list, and cleaned sentences are stored separately.
Finally, the function returns both tokens and cleaned sentences in a structured format.

### **Task 7: Error Handling**

```python
def preprocess_text(text):
    # Handle empty or invalid input
    if not text or not isinstance(text, str):
        return [], ""

    import re

    # Remove URLs and emails
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters (this also removes emojis)
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    # Remove short tokens (≤2), keep "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
test_cases = [
    "",                # Empty string
    "😍😍🔥🔥",        # Only emojis
    "1234567890",      # Only numbers
]

for case in test_cases:
    tokens, cleaned = preprocess_text(case)
    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-" * 40)
```

The preprocessing function is designed to handle different edge cases safely.
If the input is an empty string, it returns empty tokens and an empty sentence without errors.                                             
If the input contains only emojis or special characters, they are removed during preprocessing, resulting in clean empty output.
Similarly, if the input contains only numbers, they are removed, and the function returns an empty result.                                    
This ensures that the program does not crash and works reliably for all types of inputs.
